```python
ph_coal_outlet = pd.read_csv("pyro process ph coal vs outlet.csv")
ph_coal_outlet = ph_coal_outlet.iloc[:8785,1:]
```

```python
import matplotlib.pyplot as plt
ph_coal_outlet = ph_coal_outlet.to_numpy()
feed_150 = []
for o in range(len(ph_coal_outlet)):
    if ph_coal_outlet[o][0]+ph_coal_outlet[o][1]<150:
        feed_150.append(o)
```

```python
ph_coal_outlet = np.delete(ph_coal_outlet,feed_150,0)
print(ph_coal_outlet.shape)
```
