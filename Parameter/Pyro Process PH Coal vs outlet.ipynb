```python
import numpy as np
import pandas as pd
```

```python
kiln = pd.read_csv("KILN PERFORMANCE.csv")
```

```python
kiln.head(11)
```

```python
kiln = kiln.drop([0,1,2,3,4,5,6])
```

```python
kiln = kiln.iloc[:,[0,1,2,3,4,5,8,9,10,11,14]]
```

```python
pre_heater_material = pd.read_csv("pre heater gas and material temp.csv")
```

```python
pre_heater_material.head(11)
```

```python
pre_heater_material = pre_heater_material.drop([0,1,2,3,4,5,6])
```

```python
pre_heater_material = pre_heater_material.iloc[:,3:]
```

```python
pre_heater_pressure = pd.read_csv("pre heater draft profile.csv")
```

```python
pre_heater_pressure.head(15)
```

```python
pre_heater_pressure = pre_heater_pressure.drop([0,1,2,3,4,5,6])
```

```python
pre_heater_pressure = pre_heater_pressure.iloc[:,[12,13,14]]
pre_heater_pressure.head(11)
```

```python
pre_heater_fan = pd.read_csv("pre heater fan june 2019.csv")
```

```python
pre_heater_fan.head(11)
```

```python
pre_heater_fan = pre_heater_fan.drop([0,1,2,3,4,5,6])
bag = pd.read_csv("BAG FILTER.csv")
bag = bag.drop([0,1,2,3,4,5,6])
bag = bag.iloc[:,9]
```

```python
pre_heater_fan = pre_heater_fan.iloc[:,[5,12]]
pre_heater_fan.head(11)
```

```python
ph_coal_outlet = pd.concat([kiln,pre_heater_material,pre_heater_pressure,bag,pre_heater_fan],axis=1)
```

```python
ph_coal_outlet.head(330)
```

```python
feed_150 = []
for o in range(len(ph_coal_outlet)):
    if o>=3 and float(ph_coal_outlet.iloc[o,1])+float(ph_coal_outlet.iloc[o,2])<150:
        feed_150.append(o)
print(feed_150)
```

```python
for o in range(len(feed_150)):
    feed_150[o]=feed_150[o]+7
ph_coal_outlet = ph_coal_outlet.drop(feed_150)
ph_coal_outlet.to_csv("ph coal outlet.csv")
```
